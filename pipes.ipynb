```r
library(dplyr)
data("ToothGrowth")
ToothGrowth
```

```r
filtered_tg <- filter(ToothGrowth, dose == 0.5)
filtered_tg
```

```r
arrange(filtered_tg, len)
```

```r
arrange(filter(ToothGrowth, dose == 0.5), len)
```

```r
filtered_toothgrowth <- ToothGrowth  %>% 
    filter(dose == 0.5)  %>% 
    arrange(len)
filtered_toothgrowth
#View() is not supported yet
```

```r
filtered_toothgrowth <- ToothGrowth  %>% 
    filter(dose == 0.5)  %>% 
    group_by(supp)  %>% 
    summarize(mean_len = mean(len,na.rm = T),.group="drop")
filtered_toothgrowth
```

```r
library(ggplot2)
ggplot(data=ToothGrowth, aes(x=dose, y = len)) + geom_point(aes(color = supp)) + 
    labs(title = "Tooth growth: Tooth Length vs. Dosage")
```
